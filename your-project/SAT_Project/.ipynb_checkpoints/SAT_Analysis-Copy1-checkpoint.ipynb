**Does the composition of the class affect performance in tests?**

Import all relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Import the data

```python
sat_scores=pd.read_csv("./Data/2012_SAT_Results.csv")
class_sizes=pd.read_csv("./Data/2010-2011_Class_Size.csv")
demographics=pd.read_csv("./Data/School_Demographics.csv")
```

**Data Cleaning**

There are no missing values in the sat_scores dataset

```python
sat_scores.isnull().sum()
```

But the class_sizes dataset has many missing values

```python
class_sizes.isnull().sum()
```

```python
demographics.isnull().sum()
```

Construct DBN field and add it into the class_sizes dataset

```python
class_sizes["DBN"] = class_sizes.apply(lambda x: "{0:02d}{1}".format(x["CSD"], x["SCHOOL CODE"]), axis=1)
```

Only select values from class_size where GRADE is 09-12.


```python
class_av = class_sizes[class_sizes["GRADE "] == "09-12"]
class_av= class_av[class_av["PROGRAM TYPE"] == "GEN ED"]
class_av = class_av.groupby("DBN").agg(np.mean)
class_av.reset_index(inplace=True)
class_av=pd.DataFrame(class_av)
```

The demographics dataset had data for many years. So just keep the year you need.

```python
demographics.head(10)
demographics= demographics[demographics["schoolyear"] == 20112012]
```

```python
demographics.head(10)
```

```python
pd.DataFrame(demographics.columns.values)
```

Get the data for class sizes

```python
class_av
```

```python
subset_class_av=class_av[["AVERAGE CLASS SIZE", "DBN"]]
```

```python
subset_class_av.shape
```

```python
sat_scores.shape
```

Merge the column with info about the class size

```python
data1=pd.merge(sat_scores, subset_class_av, on='DBN')
```

```python
data1.shape
```

Merge demographic data (Only the one you need. Since SAT is for highschool students drop until grade 8)

```python
data2=pd.merge(data1, demographics, on='DBN')
```

```python
data2.shape
```

Transform the data into the appropriate type to be able to operate with them

```python
data2.dtypes
```

```python
#sat_scores["SAT Critical Reading Avg. Score"].astype(int)
```

```python
data2['SAT Math Avg. Score']=data2['SAT Math Avg. Score'].convert_objects(convert_numeric=True)
data2['SAT Writing Avg. Score']=data2['SAT Writing Avg. Score'].convert_objects(convert_numeric=True)
data2['SAT Critical Reading Avg. Score']=data2['SAT Critical Reading Avg. Score'].convert_objects(convert_numeric=True)
```

Check that the scores are now converted to floats

```python
data2.dtypes
```

Create a new column with the global scores for the SAT

```python
data2["Total_score_Avg"]=data2["SAT Critical Reading Avg. Score"]+data2["SAT Math Avg. Score"]+data2["SAT Writing Avg. Score"]
```
